```python
#This notebook demonstrates basic plotting using **Matplotlib** including line plots, bar charts, pie charts, and histograms.
#Simple Line Plot
#We plot a line graph with x and y data, add a title, axis labels, and markers.
```

```python
import matplotlib.pyplot as plt
```

```python
#define the Data
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
```

```python
#create line plot
plt.plot(x, y, marker='o', linestyle='-', color='b')
```

```python
# Add LAbels and titles
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Simple line plot")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
```

```python
# Create a pie chart to represent the percentage distribution of a companys revenue from different regions  
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
```

```python
# Create pie chart with labels
plt.pie(revenue, labels=regions, autopct='%1.1f%%', startangle=90)
#Add title 
plt.title("Company revenue distributions by regions")
plt.axis('equal')
plt.show()
```

```python
#Data
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
```

```python
# Index of the region with the highest revenue
max_index = revenue.index(max(revenue))
explode = [0.1 if i == max_index else 0 for i in range(len(revenue))]
plt.pie(revenue, labels=regions, autopct='%1.1f%%', startangle=90, explode=explode, shadow=True, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title("Company Revenue Distribution by Region")
plt.axis('equal')
plt.show()
```

```python
#1000 random integers between 1 and 100
import matplotlib.pyplot as plt
import random
```

```python
data = [random.randint(1, 100) for _ in range(1000)]
```

```python
# Histogram
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.title("Frequency Distribution of Random Integers (1-100)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
